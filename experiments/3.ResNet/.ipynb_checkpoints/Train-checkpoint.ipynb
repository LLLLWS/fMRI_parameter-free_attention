```python
import os
import torch
import import_ipynb
import numpy as np
from copy import deepcopy
from Network import Network
from DataInput import getHCPDataSet
from tensorboardX import SummaryWriter
import matplotlib.pyplot as plt
from ResultCalculator import resultCalculator
from LoggerCreator import loggerCreator
from EarlyStopping import EarlyStopping

# os.system('rm -rf ./logs/*')
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
logger = loggerCreator()
```

```python
HCPDataListPath = './s2_con1_label.txt'
HCPDataRootPath = '../../data/disk/'

BATCHSIZE = 16
LEARNINGRATE = 0.0001
TRAINNUM = 300
EXPERIMENTTIMES = 10
PATIENCE = 7
```

```python
dataSet = getHCPDataSet(HCPDataRootPath, HCPDataListPath, evalRate=0.35, testRate=0.1, tiny_data=100)
```

```python
def writeAcc(idx, loss, predict, trainLabel, writer, prefix=None, outputPath=None):
    acc, result = resultCalculator(predict, trainLabel, prefix=prefix, outputPath=outputPath)
    logger.info("{}: ({}) acc:{}%, loss:{}".format(idx, prefix, acc, loss))
    
    writer.add_scalar('acc', acc, idx)
    writer.add_scalar('loss', loss, idx)
    
    resultSum = result.sum(1)
    for i in range(result.shape[0]):
        writer.add_scalar('tasks/{}'.format(i), result[i][i]/resultSum[i] * 100, idx)
```

```python
def runAnEpoch(dataSet, net, lossFunc, BATCHSIZE):
    nowEpoch = dataSet.getEpoch()
    epoch = deepcopy(nowEpoch)
    loss = 0
    labels = None
    predicts = None
    
    while epoch == nowEpoch:
        data, label, epoch = dataSet.nextBatch(BATCHSIZE)
        labels = label if loss == 0 else np.hstack((labels, label))
        
        data = torch.from_numpy(data).cuda()
        label = torch.from_numpy(label).cuda()
        predict = net(data)
        
        predicts = predict.cpu().detach().numpy() if loss == 0 else np.vstack((predicts, predict.cpu().detach().numpy()))
        loss += lossFunc(predict, label).item()
        
        logger.info("Evaluationg, {}/{}".format(labels.shape[0], dataSet.getTotal()))
    return  predicts, labels, loss/labels.shape[0]
```

```python
def trainNetwork(learningRate, BatchSize, trainNum, experimentTimes = 0):
    logger.info("--------------------------------------------------------------")
    logger.info("Network loading, learningRate: {}, BatchSize: {}, trainNum: {}, experimentTimes: {}"
                .format(learningRate, BatchSize, trainNum, experimentTimes))
    trainWriter = SummaryWriter('logs/train_{}_{}_{}/'
                                .format(BatchSize, learningRate, experimentTimes))
    evalWriter = SummaryWriter('logs/eval_{}_{}_{}/'
                               .format(BatchSize, learningRate, experimentTimes))
    testWriter = SummaryWriter('logs/test_{}_{}_{}/'
                               .format(BatchSize, learningRate, experimentTimes))
    
    net = Network().cuda()
    logger.info("Network have been established.")

    lossFunc = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(net.parameters(), lr=LEARNINGRATE)
    early_stopping = EarlyStopping(PATIENCE, verbose=True, experimentTimes=experimentTimes)
    
    try:
        for idx in range(TRAINNUM):
            trainData, trainLabel, epoch = dataSet.trainDataSet.nextBatch(BatchSize)
            trainData = torch.from_numpy(trainData).cuda()
            trainLabel = torch.from_numpy(trainLabel).cuda()

            predict = net(trainData)
            loss = lossFunc(predict, trainLabel)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            if idx % 10 == 0:
                writeAcc(idx, loss.item(), predict, trainLabel, trainWriter, 'train')

            if (idx + 1) % 30 == 0:
                logger.info("Evaluating!")
                predicts, label, loss = runAnEpoch(dataSet.evalDataSet, net, lossFunc, BatchSize)
                writeAcc(idx, loss, predicts, label, evalWriter, 'eval', outputPath='./RESULT.md')

                early_stopping(loss, net)
                if early_stopping.early_stop:
                    logger.warning("Early stopping!")
                    predicts, label, loss = runAnEpoch(dataSet.testDataSet, net, lossFunc, BatchSize)
                    writeAcc(idx, loss, predicts, label, testWriter, 'eval', outputPath='./testRESULT.md')
                    logger.info("{}: model ./model/net_{}_{}_{}.pth have been saved!".format(idx, BatchSize, learningRate, experimentTimes))
                    torch.save(net, './model/net_{}_{}.pth'.format(BatchSize, learningRate, experimentTimes))
                    break

            if (idx + 1) % 100 == 0:
                logger.info("{}: model ./model/net_{}_{}_{}.pth have been saved!".format(idx, BatchSize, learningRate, experimentTimes))
                torch.save(net, './model/net_{}_{}_{}.pth'.format(BatchSize, learningRate, experimentTimes))

    except Exception as exc:
        logger.error(str(exc))
    
    if not early_stopping.early_stop:
        logger.info("{}: model ./model/net_{}_{}_{}.pth have been saved!".format(trainNum, BatchSize, learningRate, experimentTimes))
        torch.save(net, './model/net_{}_{}_{}.pth'.format(BatchSize, learningRate, experimentTimes))
        
    trainWriter.close()
    evalWriter.close()
```

```python
if __name__ == "__main__":
    for experimentTimes in range(EXPERIMENTTIMES):
        for BATCHSIZE in [16, 32]:
            for LEARNINGRATE in [0.001, 0.0003, 0.0001, 0.00003, 0.00001]:
                trainNetwork(LEARNINGRATE, BATCHSIZE, TRAINNUM, experimentTimes)
```
